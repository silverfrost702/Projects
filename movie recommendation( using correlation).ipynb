```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data',sep='\t',names = column_names)
```

```python
df.head()
```

```python
movie_names = pd.read_csv('Movie_Id_Titles')
movie_names.head(10)
```

```python
df = pd.merge(df,movie_names,on='item_id')   # we will merge both the data set for getting the titles with other information
```

```python
df.head()
```

```python
# mean rating for every title
df.groupby('title')['rating'].mean().sort_values(ascending = False).head()
```

```python
#count rating for every title
df.groupby('title')['rating'].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending = False))
```

```python
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count().sort_values(ascending = False))
```

```python
ratings.head()
```

```python
ratings['num of ratings'].hist(bins = 70)
```

```python
ratings['rating'].hist(bins = 70)
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
# make a matrix of userid X movie title such that each cel has the ratings of the respective movie
moviemat = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
moviemat.head()
```

```python
# grab any two movies
ratings.sort_values('num of ratings', ascending = False).head(10)
```

```python
# cheak user ratings for those 2 movies
starwars = moviemat['Star Wars (1977)']
liarliar = moviemat['Liar Liar (1997)']
```

```python
starwars.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars)
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['correlation'])
corr_starwars.dropna(inplace = True)
```

```python
corr_starwars.head()
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars.head()
```

```python
# set the treshold
corr_starwars[corr_starwars['num of ratings']>100].sort_values('correlation', ascending = False).head(10)
```

```python
# for liarliar
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns = ['correlation'])
corr_liarliar.dropna(inplace = True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
```

```python
corr_liarliar.head(10)
```

```python
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('correlation', ascending = False).head(10)
```
